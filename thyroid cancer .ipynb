```python
numerical_features = ['Age', 'TSH_Level', 'T3_Level', 'T4_Level', 'Nodule_Size']

# Compute correlation matrix
corr = df[numerical_features].corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
#plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,5))

# Histogram
n, bins, patches = plt.hist(df['Age'], bins=30, edgecolor='black')

# Apply a gradient colormap
colors = plt.cm.viridis(np.linspace(0, 1, len(patches)))
for patch, color in zip(patches, colors):
    patch.set_facecolor(color)

# Add KDE curve
sns.kdeplot(df['Age'], color="red", linewidth=2)

# Titles and labels
plt.title('Age Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)

plt.show()



```

```python
plt.figure(figsize=(6,4))

# Custom colors for clarity
colors = ["#4CAF50", "#FFC107", "#F44336"] 

ax = sns.countplot(
    x='Thyroid_Cancer_Risk',
    data=df,
    order=['Low','Medium','High'],
    palette=colors
)

# Add counts on top of bars
for p in ax.patches:
    ax.annotate(
        f'{p.get_height()}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='bottom',
        fontsize=10, fontweight='bold', color='black'
    )

# Titles and labels
plt.title('Thyroid Cancer Risk Levels', fontsize=14, fontweight='bold')
plt.xlabel('Risk Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.6)

plt.show()

```

```python

df['Gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(7,5))

# Histogram with KDE
sns.histplot(df['Nodule_Size'], bins=30, kde=True, color="skyblue", edgecolor="black")

# Mean & Median
mean_val = df['Nodule_Size'].mean()
median_val = df['Nodule_Size'].median()

plt.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='-', linewidth=2, label=f'Median = {median_val:.2f}')

# Titles & labels
plt.title('Distribution of Thyroid Nodule Sizes', fontsize=14, fontweight='bold')
plt.xlabel('Nodule Size (mm)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.show()

'''df['Nodule_Size'].hist()
plt.show()'''
```

```python

plt.figure(figsize=(7,5))

# Boxplot with custom colors
sns.boxplot(
    x="Gender", 
    y="Age", 
    data=df, 
    palette=["#4DB6AC", "#FFB74D"],  
    width=0.6,
    fliersize=0  
)

# Overlay swarmplot to show data distribution
sns.stripplot(
    x="Gender", 
    y="Age", 
    data=df, 
    color="black", 
    alpha=0.6, 
    jitter=True, 
    size=3
)


plt.title("Age Distribution by Gender", fontsize=14, fontweight='bold')
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()




```

Dataset Shape

How many patients (rows) and how many features/attributes (columns) are in the dataset.
Numerical Features (like Age, TSH level, T3, etc.)
Minimum, Maximum, Average (mean), and Standard Deviation.
Example: Average Age of patients, hormone level ranges.

Categorical Features (like Gender, Diagnosis, Cancer type)

Male vs Female count
Benign vs Malignant cases
Missing Data
Shows which columns have missing values and how many.
